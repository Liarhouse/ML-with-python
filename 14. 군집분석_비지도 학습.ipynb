# 군집분석이란?

1. 군집분석
 - 서로 유사한 정도에 따라 다수의 객체를 군집으로 나누는 작업 또는 이에 기반한 분석
 - 유사도가 높은 데이터끼리 그룹화(대표적으로 유클리드 거리계산식을 이용)
 - 계층형 클러스터링과 비계층형 클러스터링으로 분류
 - 주요 알고리즘 : k-means, hierarchical 등
<br><br>
2. 군집 분석의 특징
 - 종속변수(y)가 없는 데이터 마이닝 기법(비지도 학습)
 - 유클리드 거리 기반 유사 객체 묶음(유사성 = 유클리드 거리)
 - 전체적인 데이터 구조를 파악하는데 이용
 - 분석 결과에 대한 가설 검정 없음(타당성 검증 방법 없음)
 - 계층적 군집분석(탐색적), 비계층적 군집분석(확인적)
<br><br>
3. 계층적 군집분석(hierarchical)
 - 유클리드 거리를 이용한 군집분석 방법
 - 계층적으로 군집 결과 도출
 - 탐색적 군집분석
 - 계층적 군집분석의 결과 : 덴드로그램(Dendrogram) - 표본들이 군을 형성하는 과정을 나타내는 나무 형식의 그림
 - 군집 수는 사용자가 정할 수 있음
<br><br>
4. 비계층적 군집분석(k-means)
 - 계층적 군집분석보다 속도 빠름
 - 군집의 수를 알고 있는 경우 이용
 - k는 미리 정하는 군집수
 - 확인적 군집분석
 - 계층적 군집화의 결과에 의거하여 군집 수 결정
 - 변수보다 관측대상 군집화에 많이 이용
 - 군집의 중심(cluster center)은 사용자가 정함
<br><br>
5. k-means 군집분석 알고리즘
<br>
 1) k값을 초기값으로 k개의 centroid 선정(랜덤)
 <br>
 2) 각 데이터 포인터를 가장 가까운 centroid에 할당
 <br>
 3) centroid에 할당된 모든 데이터 포인터의 중심 위치 계산(centroid 재조정)
 <br>
 4) 재조정된 centroid와 가장 가까운 데이터 포인트 할당
 <br>
 5) centroid 재조정이 발생되지 않을때까지 3), 4) 단계 반복

# hierachical_계층적 군집분석
- 유클리드 거리계산식 이용

```python
from sklearn.datasets import load_iris
# 계층적 군집분석 함수
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
iris = load_iris()
```

```python
iris.feature_names
```

```python
cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
iris_df = pd.DataFrame(iris.data, columns = [cols])
```

```python
X, y = load_iris(True)
```

## 계층적 군집분석

```python
# 계층적군집분석의 연결방식 : 단순연결(single), 완전연결(complete), 평균연결방식(average)
cluster = linkage(y=iris_df, method='complete', metric='euclidean')
cluster.shape
```

```python
# 군집수 결정 : 덴드로그램 시각화
plt.figure(figsize=(25, 10))
# leaf_rotation=90 글자 각도
# leaf_font_size=12 글자크기
dendrogram(cluster, leaf_rotation=90, leaf_font_size=12)
plt.show()
```

```python
# 클러스터링 결과
from scipy.cluster.hierarchy import fcluster # 지정한 클러스터의 수로 자루는 함수

cut_tree = fcluster(cluster, t=4, criterion = 'distance')

# prediction
cut_tree 
```

```python
labels = iris.target #정답
df = pd.DataFrame({'pred':cut_tree, 'labels':iris.target})
df
```

```python
con_mat = pd.crosstab(df['pred'], df['labels'])
con_mat
```

```python
iris_df.head()
```

```python
iris_df['cluster'] = cut_tree
iris_df.head()
```

```python
# 클러스터로 분리된 값 산점도 시각화
plt.scatter(x=iris_df['sepal_length'], y=iris_df['petal_length'], c=iris_df['cluster'])
plt.show()
```

```python
# 클러스터 빈도수
iris_df['cluster'].value_counts()
```

```python
# 각클러스터별 통계
cluster_g = iris_df.groupby('cluster').mean()
cluster_g
```

# k-means : 비계층적 군집분석

```python
iris_df.head()
```

```python
# target을 제외한 feature를 이용해서 데이터 프레임을 생성
iris = load_iris()
cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
iris_df = pd.DataFrame(data=iris.data, columns=cols)
```

```python
iris_df.head()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 3, init='k-means++', max_iter=300, random_state=0)
kmeans.fit(iris_df)
```

```python
kmeans.labels_
```

```python
iris_df['target'] = iris.target
iris_df['cluster'] = kmeans.labels_

iris_result = iris_df.groupby(['target', 'cluster']).size()
iris_result
```

```python
from sklearn.decomposition import PCA

# PCA 차원축소
pca = PCA(n_components = 2)
pca_transformed = pca.fit_transform(iris.data)

iris_df['pca_x'] = pca_transformed[:, 0]
iris_df['pca_y'] = pca_transformed[:, 1]

iris_df.head()
```

```python
sns.scatterplot(x='pca_x', y = 'pca_y', data = iris_df, hue = 'cluster', style='cluster')
plt.show()
```

```python
# 클러스터링 결과
pred = kmeans.predict(iris_df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
pred
```

```python
len(pred)
```

```python
# 군집결과 시각화
plt.scatter(x=iris_df['sepal_length'], y = iris_df['petal_length'], c=pred, cmap='winter')
plt.show()
```

```python
# 군집별 중앙값
centers = kmeans.cluster_centers_
centers
```

```python
# 군집별 중앙값 시각화
plt.scatter(x=centers[:, 0], y = centers[:, 2], marker = '^', c='r')
plt.show()
```

```python
plt.scatter(x=iris_df['sepal_length'], y = iris_df['petal_length'], c=pred, cmap='winter')
plt.scatter(x=centers[:, 0], y = centers[:, 2], marker = '^', c='r')
plt.show()
```
